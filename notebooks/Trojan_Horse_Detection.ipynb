```python
def plots(ims, figsize=(20, 30), rows=10, interp=False, titles=None):
    """Display a grid of malware images with class labels."""
    if type(ims[0]) is np.ndarray:
        ims = np.array(ims).astype(np.uint8)
        if ims.shape[-1] != 3:
            ims = ims.transpose((0, 2, 3, 1))
    f    = plt.figure(figsize=figsize)
    cols = 10
    class_keys = list(batches.class_indices.keys())
    for i in range(min(50, len(ims))):
        sp = f.add_subplot(rows, cols, i + 1)
        sp.axis('Off')
        if titles is not None:
            sp.set_title(class_keys[np.argmax(titles[i])], fontsize=8)
        plt.imshow(ims[i], interpolation=None if interp else 'none')
    plt.suptitle('Sample Malware Images (MalImg Dataset)', fontsize=14)
    plt.tight_layout()
    plt.show()

plots(imgs, titles=labels)
```

## 4. Module 1 — CNN Classification Model (Keras)

### Architecture
- `Conv2D(30) → MaxPool → Conv2D(15) → MaxPool → Dropout(0.25)`
- `Flatten → Dense(128, relu) → Dropout(0.5) → Dense(50, relu) → Dense(num_classes, softmax)`
- Optimizer: **Adam** | Loss: **Categorical Cross-Entropy** | Input: **64×64 RGB**

```python
# ── Train / Test split ────────────────────────────────────────────────────────
X_train, X_test, y_train, y_test = train_test_split(
    imgs / 255., labels, test_size=0.2, random_state=42
)
num_classes = labels.shape[1]
print(f"X_train : {X_train.shape} | X_test : {X_test.shape}")
print(f"y_train : {y_train.shape} | y_test : {y_test.shape}")
print(f"Number of classes : {num_classes}")
```

```python
def malware_model():
    """Build and compile the Keras CNN model for malware classification."""
    model = Sequential([
        Conv2D(30, kernel_size=(3, 3), activation='relu', input_shape=(64, 64, 3)),
        MaxPooling2D(pool_size=(2, 2)),
        Conv2D(15, (3, 3), activation='relu'),
        MaxPooling2D(pool_size=(2, 2)),
        Dropout(0.25),
        Flatten(),
        Dense(128, activation='relu'),
        Dropout(0.5),
        Dense(50, activation='relu'),
        Dense(num_classes, activation='softmax'),
    ])
    model.compile(
        loss='categorical_crossentropy',
        optimizer='adam',
        metrics=['accuracy']
    )
    return model

Malware_model = malware_model()
Malware_model.summary()
```

```python
# ── Compute class weights to handle imbalance ─────────────────────────────────
y_train_new       = np.argmax(y_train, axis=1)
class_weights     = compute_class_weight(
    class_weight='balanced',
    classes=np.unique(y_train_new),
    y=y_train_new
)
class_weights_dict = dict(zip(np.unique(y_train_new), class_weights))

# ── Train ─────────────────────────────────────────────────────────────────────
history_keras = Malware_model.fit(
    X_train, y_train,
    validation_data=(X_test, y_test),
    epochs=10,
    class_weight=class_weights_dict,
    verbose=1
)

scores = Malware_model.evaluate(X_test, y_test, verbose=0)
print(f"\n✅ Final CNN (Keras) Accuracy : {scores[1]*100:.2f}%")
```

```python
# ── Plot training history ─────────────────────────────────────────────────────
fig, axes = plt.subplots(1, 2, figsize=(14, 4))

axes[0].plot(history_keras.history['loss'],     label='Train Loss')
axes[0].plot(history_keras.history['val_loss'], label='Val Loss')
axes[0].set_title('Loss Curve')
axes[0].set_xlabel('Epoch')
axes[0].legend()

axes[1].plot(history_keras.history['accuracy'],     label='Train Acc')
axes[1].plot(history_keras.history['val_accuracy'], label='Val Acc')
axes[1].set_title('Accuracy Curve')
axes[1].set_xlabel('Epoch')
axes[1].legend()

plt.suptitle('Module 1 — Keras CNN Training History', fontsize=13)
plt.tight_layout()
plt.show()
```

```python
# ── Confusion Matrix (Keras model) ────────────────────────────────────────────
y_pred_prob = Malware_model.predict(X_test)
y_pred      = np.argmax(y_pred_prob, axis=1)
y_test2     = np.argmax(y_test,      axis=1)

c_matrix    = metrics.confusion_matrix(y_test2, y_pred)
class_names = list(batches.class_indices.keys())

def plot_confusion_matrix(confusion_matrix, class_names, figsize=(20, 7), fontsize=14):
    df_cm = pd.DataFrame(confusion_matrix, index=class_names, columns=class_names)
    fig   = plt.figure(figsize=figsize)
    heatmap = sns.heatmap(df_cm, annot=True, fmt='d', cmap='Blues')
    heatmap.yaxis.set_ticklabels(heatmap.yaxis.get_ticklabels(), rotation=0,  ha='right',  fontsize=fontsize)
    heatmap.xaxis.set_ticklabels(heatmap.xaxis.get_ticklabels(), rotation=45, ha='right', fontsize=fontsize)
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    plt.title('Confusion Matrix — Module 1 (Keras CNN Classification)', fontsize=15)
    plt.tight_layout()
    plt.show()

plot_confusion_matrix(c_matrix, class_names, figsize=(20, 7), fontsize=10)
```

## 5. Module 2 — CNN Detection Model (PyTorch)

### Architecture (Net class)
Five convolutional layers (3→16→32→64→128→64 filters) each followed by MaxPooling, then three fully connected layers with Dropout(0.5). Output uses `log_softmax` for binary classification (benign / malware).

```python
# ── Data transforms & loaders ────────────────────────────────────────────────
transform = transforms.Compose([
    transforms.Resize((256, 256)),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406],
                         [0.229, 0.224, 0.225]),
])

train_ds = datasets.ImageFolder(train_path, transform=transform)
val_ds   = datasets.ImageFolder(val_path,   transform=transform)

BATCH_SIZE = 64
train_set  = torch.utils.data.DataLoader(train_ds, batch_size=BATCH_SIZE, shuffle=True)
val_set    = torch.utils.data.DataLoader(val_ds,   batch_size=BATCH_SIZE, shuffle=True)

classes = os.listdir(train_path)
print(f"Classes : {classes}")
print(f"Train batches : {len(train_set)} | Val batches : {len(val_set)}")
```

```python
def imshow(image, title=None):
    """Unnormalise and display a single tensor image."""
    image = image.numpy().transpose((1, 2, 0))
    mean  = np.array([0.485, 0.456, 0.406])
    std   = np.array([0.229, 0.224, 0.225])
    image = np.clip(image * std + mean, 0, 1)
    plt.imshow(image)
    if title:
        plt.title(title)
    plt.axis('off')
    plt.pause(0.001)

# Show a sample batch
images_sample, labels_sample = next(iter(train_set))
grid = torchvision.utils.make_grid(images_sample[:16])
plt.figure(figsize=(14, 4))
imshow(grid, title=[classes[x] for x in labels_sample[:16]])
plt.show()
```

```python
NUM_CLASSES = 2  # benign / malware

class Net(nn.Module):
    """5-layer CNN for binary malware detection."""

    def __init__(self):
        super().__init__()
        # Convolutional layers
        self.conv1 = nn.Conv2d(3,   16,  3, padding=1)
        self.conv2 = nn.Conv2d(16,  32,  3, padding=1)
        self.conv3 = nn.Conv2d(32,  64,  3, padding=1)
        self.conv4 = nn.Conv2d(64,  128, 3, padding=1)
        self.conv5 = nn.Conv2d(128, 64,  3, padding=1)
        # Pooling
        self.pool  = nn.MaxPool2d(2, 2)
        # Fully connected  (256 × 4 × 4 = 4096 after 5× pool on 256-input)
        self.fc1     = nn.Linear(256 * 4 * 4, 1024)
        self.fc2     = nn.Linear(1024, 1024)
        self.fc3     = nn.Linear(1024, NUM_CLASSES)
        self.dropout = nn.Dropout(p=0.5)

    def forward(self, x):
        x = self.pool(F.relu(self.conv1(x)))
        x = self.pool(F.relu(self.conv2(x)))
        x = self.pool(F.relu(self.conv3(x)))
        x = self.pool(F.relu(self.conv4(x)))
        x = self.pool(F.relu(self.conv5(x)))
        x = x.view(-1, 256 * 4 * 4)
        x = self.dropout(F.relu(self.fc1(x)))
        x = self.dropout(F.relu(self.fc2(x)))
        x = F.log_softmax(self.fc3(x), dim=1)
        return x


model = Net().to(device)
summary(model, input_size=(3, 256, 256))
```

## 6. Hyperparameter Tuning with TensorBoard

```python
import shutil
if os.path.exists('./runs/'):
    shutil.rmtree('./runs/')

torch.autograd.set_detect_anomaly(True)

parameters = dict(
    dropout   = [0.1, 0.3, 0.5],
    lr        = [0.01, 0.001],
    batch_size= [32, 64, 128],
    shuffle   = [True, False],
)
param_val = [v for v in parameters.values()]


def get_num_correct(preds, labels):
    return preds.argmax(dim=1).eq(labels).sum().item()


def filter_valid_samples(preds, labels, num_classes):
    """Remove out-of-bounds labels that can arise from data issues."""
    valid_mask   = (labels >= 0) & (labels < num_classes)
    valid_preds  = preds[valid_mask]
    valid_labels = labels[valid_mask]
    valid_labels[valid_labels < 0]           = 0
    valid_labels[valid_labels >= num_classes] = 0
    valid_preds_mask = torch.any(valid_preds >= 0, dim=1)
    return valid_preds[valid_preds_mask], valid_labels[valid_preds_mask]


print("Starting hyperparameter search (1 epoch each combination for demo) …")
for hyperparam_test_id, (dropout, lr, batch_size, shuffle) in enumerate(product(*param_val)):
    if hyperparam_test_id >= 4:   # limit to 4 runs in this notebook demo
        break
    print(f"  Run {hyperparam_test_id+1}: dropout={dropout} lr={lr} batch={batch_size} shuffle={shuffle}")

    model_hp   = Net().to(device)
    loader_hp  = torch.utils.data.DataLoader(train_ds, batch_size=batch_size, shuffle=shuffle)
    optimizer_hp  = optim.Adam(model_hp.parameters(), lr=lr)
    criterion_hp  = nn.CrossEntropyLoss(ignore_index=-1)
    comment       = f'dropout={dropout} batch={batch_size} lr={lr} shuffle={shuffle}'
    writer_hp     = SummaryWriter(comment=comment)

    total_loss, total_correct = 0, 0
    for images_hp, labels_hp in loader_hp:
        images_hp, labels_hp = images_hp.to(device), labels_hp.to(device)
        preds_hp = model_hp(images_hp)
        vp, vl   = filter_valid_samples(preds_hp, labels_hp, NUM_CLASSES)
        loss_hp  = criterion_hp(vp, vl)
        total_loss    += loss_hp.item()
        total_correct += get_num_correct(vp, vl)
        optimizer_hp.zero_grad()
        loss_hp.backward()
        optimizer_hp.step()

    writer_hp.add_hparams(
        {'dropout': dropout, 'lr': lr, 'batch_size': batch_size, 'shuffle': shuffle},
        {'accuracy': total_correct / len(loader_hp), 'loss': total_loss}
    )
    writer_hp.close()

print("\n✅ Hyperparameter sweep complete.")
```

## 7. Full Training Loop with Validation

```python
# ── Re-initialise model, criterion, optimiser ─────────────────────────────────
model     = Net().to(device)
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)
scheduler = lr_scheduler.StepLR(optimizer, step_size=7, gamma=0.1)

NUM_EPOCHS = 10     # increase for production; 1 epoch shown in demo
history = {'train_loss': [], 'valid_loss': [], 'train_acc': [], 'valid_acc': []}

writer_main = SummaryWriter()

for epoch in range(NUM_EPOCHS):
    # ── Training phase ────────────────────────────────────────────────────────
    model.train()
    train_loss, train_correct = 0.0, 0
    for images, labels in train_set:
        images, labels = images.to(device), labels.to(device)
        optimizer.zero_grad()
        output = model(images)
        loss   = criterion(output, labels)
        loss.backward()
        optimizer.step()
        train_loss    += loss.item() * images.size(0)
        _, predictions = torch.max(output.data, 1)
        train_correct += (predictions == labels).sum().item()

    # ── Validation phase ──────────────────────────────────────────────────────
    model.eval()
    valid_loss, val_correct = 0.0, 0
    with torch.no_grad():
        for images, labels in val_set:
            images, labels = images.to(device), labels.to(device)
            output = model(images)
            loss   = criterion(output, labels)
            valid_loss  += loss.item() * images.size(0)
            _, predictions = torch.max(output.data, 1)
            val_correct += (predictions == labels).sum().item()

    scheduler.step()

    train_loss_avg = train_loss / len(train_set.sampler)
    train_acc_avg  = train_correct / len(train_set.sampler) * 100
    valid_loss_avg = valid_loss / len(val_set.sampler)
    valid_acc_avg  = val_correct / len(val_set.sampler) * 100

    history['train_loss'].append(train_loss_avg)
    history['valid_loss'].append(valid_loss_avg)
    history['train_acc'].append(train_acc_avg)
    history['valid_acc'].append(valid_acc_avg)

    writer_main.add_scalar('Loss/Train',      train_loss_avg, epoch)
    writer_main.add_scalar('Loss/Validation', valid_loss_avg, epoch)
    writer_main.add_scalar('Acc/Train',       train_acc_avg,  epoch)
    writer_main.add_scalar('Acc/Validation',  valid_acc_avg,  epoch)

    print(f"Epoch [{epoch+1:02d}/{NUM_EPOCHS}]  "
          f"Train Loss: {train_loss_avg:.4f}  Train Acc: {train_acc_avg:.2f}%  "
          f"Val Loss: {valid_loss_avg:.4f}  Val Acc: {valid_acc_avg:.2f}%")

writer_main.close()
print("\n✅ Training complete.")
```

## 8. Confusion Matrix Visualization (PyTorch CNN)

```python
nb_classes       = NUM_CLASSES
confusion_matrix = torch.zeros(nb_classes, nb_classes)

model.eval()
with torch.no_grad():
    for data in val_set:
        images, labels = data
        images, labels = images.to(device), labels.to(device)
        outputs        = model(images)
        _, preds       = torch.max(outputs, 1)
        for t, p in zip(labels.view(-1), preds.view(-1)):
            confusion_matrix[t.long(), p.long()] += 1

cm_norm     = confusion_matrix.numpy()
cm_norm     = cm_norm / cm_norm.sum(axis=1, keepdims=True)
class_names = ('benign', 'malware')

fig = plt.figure(figsize=(8, 6))
sns.heatmap(
    cm_norm, annot=True, cmap='gray',
    xticklabels=class_names, yticklabels=class_names,
    fmt='.2f'
)
plt.ylabel('True Label')
plt.xlabel('Predicted Label')
plt.title('Confusion Matrix — Module 2 (PyTorch CNN Detection)', fontsize=13)
plt.tight_layout()
plt.show()

# Log to TensorBoard
writer_cm = SummaryWriter()
writer_cm.add_figure('Confusion Matrix', fig)
writer_cm.close()
```

## 9. Random Forest Classifier for Accuracy Evaluation

CNN features are extracted from the penultimate convolutional layer and fed into a **Random Forest** (100 trees) to compute final detection accuracy — as described in the project algorithm (Step 5).

```python
def extract_features(images_tensor):
    """Extract CNN feature maps from conv5 layer."""
    with torch.no_grad():
        features = model.conv5(
            model.pool(model.conv4(
                model.pool(model.conv3(
                    model.pool(model.conv2(
                        model.pool(model.conv1(images_tensor))
                    ))
                ))
            ))
        )
    return features.view(features.size(0), -1)


# ── Extract features from full training set ───────────────────────────────────
all_features, all_labels = [], []
model.eval()
for images_rf, labels_rf in train_set:
    feats = extract_features(images_rf.to(device))
    all_features.extend(feats.cpu().numpy())
    all_labels.extend(labels_rf.cpu().numpy())

# ── Train / test split ────────────────────────────────────────────────────────
X_train_rf, X_test_rf, y_train_rf, y_test_rf = train_test_split(
    all_features, all_labels, test_size=0.2, random_state=42
)

# ── Fit Random Forest ─────────────────────────────────────────────────────────
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42, n_jobs=-1)
rf_classifier.fit(X_train_rf, y_train_rf)

# ── Evaluate ──────────────────────────────────────────────────────────────────
accuracy = rf_classifier.score(X_test_rf, y_test_rf)
print(f"✅ Random Forest Accuracy : {accuracy*100:.2f}%")
```

```python
# ── Random Forest Confusion Matrix ────────────────────────────────────────────
y_pred_rf = rf_classifier.predict(X_test_rf)
cm_rf     = metrics.confusion_matrix(y_test_rf, y_pred_rf)

fig_rf = plt.figure(figsize=(6, 5))
sns.heatmap(
    cm_rf, annot=True, fmt='d', cmap='Blues',
    xticklabels=['Benign', 'Malware'],
    yticklabels=['Benign', 'Malware']
)
plt.ylabel('True Label')
plt.xlabel('Predicted Label')
plt.title('Confusion Matrix — Random Forest (Step 5 Accuracy)', fontsize=12)
plt.tight_layout()
plt.show()

print(metrics.classification_report(y_test_rf, y_pred_rf, target_names=['Benign', 'Malware']))
```
